```python
import chromadb
from chromadb.utils import embedding_functions
from sentence_transformers import SentenceTransformer
```

```python
client = chromadb.Client()
```

```python
collection = client.create_collection(name="first_collection")
```

Using default embedding model (all-MiniLM-L6-v2 ) provided by chroma

```python

collection.add(
    ids= ["1","2","3"],
    documents =["I am learning chromadb.",
                "chromadb is used in rag systems",
                "embeddings turn text to numbers"])
results = collection.query(
    query_texts = ["rag and chroma"],
    n_results= 1
)
print(results)
```

```python

model = SentenceTransformer(model_name_or_path='all-MiniLM-L6-v2')
text = ["I am learning chromadb.",
                "chromadb is used in rag systems",
                "embeddings turn text to numbers"]
embeddings= model.encode(text)
collection.add(
                ids= ["1","2","3"],
                documents = text,
                embeddings=embeddings
    
)
collection.query(
    query_embeddings=model.encode("what are he learning?"),
    n_results=1
)
```

Using embedding function

```python
embedding_fnc= embedding_functions.SentenceTransformerEmbeddingFunction(model_name="all-MiniLM-L6-v2")
collection=client.create_collection(
    name='second_collection',
    embedding_function=embedding_fnc
)
collection.add(
    ids=["1","2","3"],
    documents =  ["I am learning chromadb.",
                "chromadb is used in rag systems",
                "embeddings turn text to numbers"],
)
result= collection.query(
    query_texts = ["what is rag?"],
    n_results= 1
)
print(result)
```

```python
search = index.search(
    namespace="namespace1",
    query={
        "top_k":10,
        "inputs":{
            'text':"famous monuments"
        }

    }

)
print(search['result']['hits'][0])
```

Reranking result using integrated reranking model bge-reranker

```python
search = index.search(
    namespace="namespace1",
    query={
        "top_k":3,
        "inputs":{
            'text':"famous monuments"
        }},
        rerank ={ "model":"bge-reranker-v2-m3",
                "top_n":2,
                "rank_fields":["chunk_text"] }
)
print(search)
```

Reranking using standalone reranking method 

```python
ranked_results = pc.inference.rerank(
    model = "bge-reranker-v2-m3",
    query= "famous monuments",
    documents= records,
    top_n= 1 ,
    rank_fields=["chunk_text"],
    return_documents=True
)
print(ranked_results)
```

Search with Metadata Filter

```python
filtered_results = index.search(
    namespace="namespace1", 
    query={
        "inputs": {"text": "historical monuments"}, 
        "top_k": 3,
        "filter": {"category": {"$eq":"history"}} ,
    },
    fields=["category", "chunk_text"]
)

print(filtered_results)
```

Lexical search using sparse index

```python
index_name = "sparse-index"
if index_name not in pc.list_indexes():
    pc.create_index(
        name=index_name,
        vector_type="sparse",
        metric='dotproduct',
        spec=ServerlessSpec(cloud='aws', region='us-east-1')
    )
```

```python
index=pc.Index(index_name)

index.upsert(
    vectors=[
        {
            "id": "d1",
            "sparse_values": {"indices": [3, 6, 9], "values": [0.3, 0.9, 0.5]},
            "metadata": {"text": "AI improves medicine"}
        },
        {
            "id": "d2",
            "sparse_values": {"indices": [2, 5, 8], "values": [0.8, 0.2, 0.7]},
            "metadata": {"text": "AI powers cars"}
        }
    ]
)
query_results = index.query(
    top_k=1,
    sparse_vector={
        "values":[0.1,0.6,0.9,1.2],
        "indices":[3,6,8,9]
    },
    include_values=True
    )
print(query_results)
```

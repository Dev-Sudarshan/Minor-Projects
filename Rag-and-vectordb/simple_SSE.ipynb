```python
from sentence_transformers import SentenceTransformer
from sklearn.metrics.pairwise import cosine_similarity
import numpy as np
```

```python
dataset= [
    "The capital of France is Paris.","The capital of Germany is Berlin.",
    "The capital of Italy is Rome.","The capital of Spain is Madrid.",
    "The capital of Portugal is Lisbon.","The capital of Netherlands is Amsterdam."]
query= "what is the capital of france?"
```

```python
model = SentenceTransformer('all-MiniLM-L6-v2')
```

```python
data_embeddings= model.encode(dataset)
query= "what is the capital of france?"
query_embedding= model.encode([query])
```

```python
similarity_scores= cosine_similarity(query_embedding, data_embeddings)
matching_idx= np.argmax(similarity_scores)
```

```python
print(f"Query: {query}")
print(f"Most similar sentence: {dataset[matching_idx]}")
#Most similar sentence: The capital of France is Paris.
```

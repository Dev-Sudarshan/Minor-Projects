```python
import torch
import torch.nn as nn 
import torch.optim as optim
from torchmetrics.classification import BinaryAccuracy
```

```python
#features= [has_wings,lays_eggs,can_fly,has_feathers,body_temp]
dataset= [[1, 1, 1, 1, 39],
    [1, 1, 1, 1, 38],
    [0, 0, 0, 0, 36],
    [0, 0, 0, 0, 36],  
    [0, 0, 0, 0, 37],  
    [1, 0, 1, 0, 37],  
    [1, 1, 0, 0, 38]]
X= torch.tensor(dataset,dtype= torch.float32)
label= [ [1], [1], [0],[0], [0], [0], [0]]
Y= torch.tensor(label , dtype = torch.float32)
```

```python
model= model = nn.Sequential(
    nn.Linear(5, 4),
    nn.ReLU(),
    nn.Linear(4, 1),
    nn.Sigmoid()
    )
```

```python
criterion= nn.BCELoss()
optimizer= optim.SGD(model.parameters(),lr=0.01)
```

```python
epoch =1000
total_loss = 0.0
for i in range(epoch):
    outputs= model(X)
    loss=criterion(outputs, Y)
    optimizer.zero_grad()
    loss.backward()
    optimizer.step()
    total_loss += loss.item()
    accuracy = BinaryAccuracy(threshold=0.5)
    accuracy.update(outputs, Y)
    accuracy_value = accuracy.compute()
    if((i+1)%100==0):
        print(f"For {i+1} epoch, loss = {loss.item()} , accuracy = {accuracy_value}")
print(f"Total loss after {epoch} epochs: {total_loss}")



```

```python
model.eval()
with torch.no_grad():
     test_samples = torch.tensor([
        [1, 1, 1, 1, 39],  
        [0, 0, 0, 0, 36]   ], dtype=torch.float32)
     predictions = model(test_samples)
     predicted_classes = (predictions >= 0.5).bool()
     print("Predicted probabilities:\n", predicted_classes)
```
